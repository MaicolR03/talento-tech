###**CALCULADORA BASICA**

```python
def sum(x,y):
  return x + y

def substract(x,y):
  return x - y

def multiply(x,y):
  return x * y

def divide(x,y):
  try:
    return x / y
  except:
    print("Error en la division")
    main(value1,value2)

def main(num1, num2):
  print("1. Suma")
  print("2. Resta")
  print("3. Multiplicacion")
  print("4. Division")


  entrada = input("Seleccione la operacion:       ")

  match entrada:
    case "1":
      suma = sum(num1,num2)
      print(suma)
    case "2":
      resta = substract(num1,num2)
      print(resta)
    case "3":
      producto = multiply(num1,num2)
      print(producto)
    case "4":
      Cociente = divide(num1,num2)
      print(Cociente)
    case _:
      print("Opcion no valida")
      main(num1,num2)

def error_handling(user):
    try:
        z = float(user)
    except:
        print("El numero digitado no es un valor numerico, por favor ingrese un valor numerico \n")
        return -1
    return z

def input_value1():
    num1 = input("Ingrese el primer numero:      ")
    validatedtype = error_handling(num1)
    if validatedtype == -1:
        input_value1()
    else:
        return validatedtype

def input_value2():
    num2 = input("Ingrese el segundo numero:      ")
    validatedtype = error_handling(num2)
    if validatedtype == -1:
        input_value2()
    else:
        return validatedtype

value1 = input_value1()
value2 = input_value2()
main(value1, value2)
```

###**TIME**

```python
from datetime import datetime  #Library for get time
import pytz  #Library for get timezones

zones = pytz.all_timezones  #Get all timezones
Bogota = pytz.timezone("America/Bogota") #Assign timezone in Bogota
time = datetime.now()  #Get time
time_bogota = time.astimezone(Bogota)  #Get time in Bogota
hour = time_bogota.strftime("%H:%M:%S") # Obtain hour in Bogota

if hour >= "00:00:00" and hour <= "12:00:00":
  print("Good Morning")
elif hour >= "12:00:01" and hour <= "18:00:00":
  print("Good Afternoon")
elif hour >= "18:00:01" and hour <= "23:59:59":
  print("Good Evening")
```
